```python
nome=soup.find_all("p",class_="product-title")[0]
print(nome.text)
```

```python
nomi=[]
nomi_soup=soup.find_all("p",class_="product-title")
for nome in nomi_soup:
    nomi.append(nome.text.strip())
print(len(nomi),'-->\n',nomi)
```

```python
# <p class="product-price">109,00 €</p>
```

```python
prezzo=soup.find_all("p",class_="product-price")[1]
print(prezzo.text)
```

```python
prezzi=[]
prezzi_soup=soup.find_all("p",class_="product-price")
for prezzo in prezzi_soup:
    prezzi.append(prezzo.text.strip())
print(len(prezzi),'-->\n',prezzi)
prezzi=prezzi[1::2]
```

```python
prezzin=[]
for prezzo in prezzi:
    prezzo=prezzo.replace(' €','')
    prezzo=prezzo.replace(',','.')
    prezzo=float(prezzo)
    prezzin.append(prezzo)
```

```python
marca=[]
for i in nomi:
    marca.append('Kammi')
```

```python
import pandas as pd
Kammi={
    "marca":marca,
    "modello":modello,
    "name":nomi,
    "prezzi":prezzin,
}
dfk=pd.DataFrame(Kammi)
dfk # è IL DATABASE DEFINITIVO DI KAMMI
```

```python
dfg=dfk.groupby("modello")
```
